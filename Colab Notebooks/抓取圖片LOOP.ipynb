```python
import requests

# 新的数字列表
# numbers = [
#     2430, 3490, 3497, 3506, 2614, 9, 727, 966, 1009, 1076, 1334, 3242, 17374
# ]

numbers = [
    18
]

# 使用这些数字生成新的 URL 列表
image_urls = [f"https://api.pudgypenguins.io/lil/image/{num}" for num in numbers]

# 打印生成的 URL 列表
# print(image_urls)
```

```python
# 遍历每个 URL 并下载图片
for url in image_urls:
    # 获取图片 ID (最后的数字部分) 作为文件名
    image_id = url.split("/")[-1]
    file_name = f"pudgy_penguin_{image_id}.png"

    # 发送 GET 请求获取图片内容
    response = requests.get(url)

    # 检查请求是否成功
    if response.status_code == 200:
        # 打开一个文件，并将内容写入文件中 (使用二进制模式)
        with open(file_name, "wb") as file:
            file.write(response.content)
        print(f"图片已成功下载并保存为 '{file_name}'")
    else:
        print(f"无法下载图片 {image_id}, 状态码: {response.status_code}")
```
