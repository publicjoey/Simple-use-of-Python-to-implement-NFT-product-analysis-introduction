執行 SQL 語句

```python
import sqlite3

# 連接到 SQLite 資料庫
conn = sqlite3.connect('NFT.db')  # 替換 'your_database.db' 為實際的資料庫名稱
cursor = conn.cursor()

# SQL 語句: UPDATE 不是使用 INSERT OR REPLACE
sql = """
DELETE FROM news3;
"""

try:
    # 執行 SQL 語句
    cursor.execute(sql)

    # 提交變更
    conn.commit()

    print("資料插入或替換成功！")
except sqlite3.IntegrityError as e:
    print(f"發生錯誤: {e}")
finally:
    # 關閉連接
    cursor.close()
    conn.close()
```

合併資料表到 traits_price

```python
import sqlite3

# 連接到 SQLite 資料庫
conn = sqlite3.connect('NFT.db')  # 替換 'your_database.db' 為實際的資料庫名稱
cursor = conn.cursor()

# SQL 語句: INSERT OR REPLACE
sql = """
INSERT OR REPLACE INTO traits_price ("nft_id","name","Left Flipper","Right Flipper","Background","Face","Skin","Head","Body")
SELECT "nft_id","name","Left Flipper","Right Flipper","Background","Face","Skin","Head","Body"
FROM news3;
"""
try:
    # 執行 SQL 語句
    cursor.execute(sql)

    # 提交變更
    conn.commit()

    print("資料插入或替換成功！")
except sqlite3.IntegrityError as e:
    print(f"發生錯誤: {e}")
finally:
    # 關閉連接
    cursor.close()
    conn.close()
```

```python
import sqlite3

# 連接到 SQLite 資料庫
conn = sqlite3.connect('NFT.db')  # 替換 'your_database.db' 為實際的資料庫名稱
cursor = conn.cursor()


# SQL 語句: UPDATE 不是使用 INSERT OR REPLACE
sql = """
UPDATE traits_price
SET price_value = (SELECT price_value FROM news2 WHERE news2.nft_id = traits_price.nft_id)
WHERE nft_id IN (SELECT nft_id FROM news2);
"""

try:
    # 執行 SQL 語句
    cursor.execute(sql)

    # 提交變更
    conn.commit()

    print("資料插入或替換成功！")
except sqlite3.IntegrityError as e:
    print(f"發生錯誤: {e}")
finally:
    # 關閉連接
    cursor.close()
    conn.close()
```
