```python

import requests
import json

# 基本的 URL，不包含特定的 NFT ID
base_url = "https://api.opensea.io/api/v2/offers/collection/lilpudgys/nfts/"
tail = "/best"

# 你的 API 密钥
headers = {
    "accept": "application/json",
    "x-api-key": APIKEYsetting
}
```

抓取區間 price

```python
# 使用 Python 生成從 1 到 99 的列表
nft_ids = list(range(start,end))
nft_ids
# 假设你要查询的 NFT ID 列表
# nft_ids = [1145, 1146, 1147]  # 你可以添加更多的 ID
# 用于存储所有 NFT 数据的列表
nft_data_list = []
```

```python
import time
import requests
import json

# 遍历每个 NFT ID 并取得资料
for nft_id in nft_ids:
    url = f"{base_url}{nft_id}{tail}"
    response = requests.get(url, headers=headers)

    # if response.status_code == 200:
    if response.status_code == 200: # 200表示请求成功
        nft_data = response.json()  # 使用 .json() 来解析 JSON 响应
        nft_data_list.append({"nft_id": nft_id, "data": nft_data})
    else:
        print(f"无法取得 NFT ID: {nft_id} 的资料, 状态码: {response.status_code}")

    # 加入间隔时间，延迟1秒
    time.sleep(speed)

# 将收集到的 NFT 数据转换为 JSON 字符串格式并输出
output_json = json.dumps(nft_data_list, indent=4, ensure_ascii=False)
# print(output_json)
```

先寫入字典, 再轉 json (修正 直接儲存 json 格式錯誤)

```python
import json

# 將 JSON 字符串轉換成 Python 字典
output_dict = json.loads(output_json)

# 確認字典內容
print(output_dict)
print(type(output_dict))
```

```python
# 存成 json file
import json
# my_dict = { 'Apple': 4, 'Banana': 2, 'Orange': 6, 'Grapes': 11}
tf = open("myDictionary.json", "w")
json.dump(output_dict,tf)
tf.close()
```

```python
import json
import csv

# 從文件中讀取 JSON 數據
with open('myDictionary.json', 'r') as file:
    data = json.load(file)

# 提取 'nft_id' 和 'price.value'
rows = []
for item in data:
    nft_id = item["nft_id"]
    price_value = item["data"]["price"]["value"]
    rows.append([nft_id, price_value])

# 保存到 CSV 文件
# csv_file = 'nft_prices_no_sqlite.csv'
csv_file = f'nft_prices_no_sqlite_{start}_{end}.csv'
with open(csv_file, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["nft_id", "price_value"])
    writer.writerows(rows)

print(f"CSV file '{csv_file}' created successfully.")
```

從 JSON 結構中提取每個 trait_type 和對應的 value，並將其轉換為字典。 使用 pandas 庫將字典列表轉換為 DataFrame 表格，方便展示或進行進一步處理。

```python
import json
with open('myDictionary_traits.json') as f:
    data = json.load(f)
```

```python
import pandas as pd
# 定義帶有 traits 的 JSON 數據
nft_data = data
```

```python
# # 假设 start 和 end 已经定义
# start = 100
# end = 105
import pandas as pd
# 定義帶有 traits 的 JSON 數據
# nft_data = [

# 從 JSON 中提取 traits 數據，並添加 "name" 欄位
traits_list_with_name = [
    {
        **{trait["trait_type"]: trait["value"] for trait in item["data"]["nft"]["traits"]},
        "name": item["data"]["nft"]["name"],
        "nft_id": item["nft_id"]
    }
    for item in nft_data
]

# 將列表轉換為 DataFrame
traits_df_with_name = pd.DataFrame(traits_list_with_name)

# 將 DataFrame 保存為 CSV 文件
traits_df_with_name.to_csv(f'traits_no_sqlite_{start}_{end}.csv', index=False)
# traits_df_with_name.to_csv('nft_prices_no_sqlite.csv', index=False)


# print("CSV 文件已保存為 traits_no_sqlite.csv")
print(f"CSV 文件已保存為 traits_no_sqlite_{start}_{end}.csv")
```
